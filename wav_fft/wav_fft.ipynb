## Test the performance Fast-Fourier Transform using SciPy on included .wav file

```python
# https://docs.scipy.org/doc/scipy/tutorial/fft.html
from scipy.fft import fft, ifft, fftfreq

#https://docs.scipy.org/doc/scipy/reference/generated/scipy.io.wavfile.read.html
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt
```

# Load data from sample WAV file

```python
# test_audio1.wav - 10s recording from 'voice memos' app
AUDIO_FILE = './audio/test_audio1.wav'
samplerate, data = wavfile.read(AUDIO_FILE)
```

```python
# normalize signal between [-1, 1]
signal = data / np.max(np.abs(data))

# perform FFT on signal
n = len(signal) # number of samples in data
amplitudes = fft(signal) # fft of signal
frequencies = fftfreq(n, 1/samplerate) # define frequency bins

# use positive frequencies only
frequencies = frequencies[:n//2]
# take absolute value of amplitudes values
amplitudes = abs(amplitudes[:n//2])

# Plot the frequency spectrum
plt.figure(figsize=(12, 8))
plt.plot(frequencies, amplitudes)
plt.title(f"Fast Fourier Transform on {AUDIO_FILE}")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```
